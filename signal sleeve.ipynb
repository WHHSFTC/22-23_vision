```python
import cv2
import numpy as np
import math as m
```

```python
lower_orange = np.array([0, 141, 0])
upper_orange = np.array([255, 255, 90])

# lower_green = np.array([0, 100, 0])
# upper_green = np.array([255, 130, 150])

lower_green = np.array([0, 100, 0])
upper_green = np.array([255, 130, 150])

lower_pink = np.array([0, 180, 125])
upper_pink = np.array([255, 255, 255])

radius = 50
start_point = (320 - 50, 240 - 50)
end_point = (320 + 50, 240 + 50)

horizon = 0
minwidth = 20

focal_length = 320/46*92.5
camera_angle = np.pi / 6.0

def ctransform(x, y):
    return (x - 320, -y + 240)
def theta(n):
    return m.atan2(n, focal_length)
def delta(n1, n2):
    return theta(n2) - theta(n1)
def estimateDistance(da):
    return 1.0 / m.sin(da/2)
```

```python
def findColor(img, label, lower, upper, conversion=cv2.COLOR_BGR2YCrCb):
    ycrcb = cv2.cvtColor(img, conversion)
    # cv2.imshow(label + " ycrcb", ycrcb)
    
    mask = cv2.inRange(ycrcb, lower, upper)
    # cv2.imshow(label + " mask", mask)
    
    mask = cv2.GaussianBlur(mask, (5, 5), cv2.BORDER_DEFAULT)
    # cv2.imshow(label + " blur", mask)
    
    res = cv2.bitwise_and(img, img, mask = mask)
    
    cv2.rectangle(res, start_point, end_point, (255, 0, 0), 2)
    
    total = 0
    n = 0
    for x in range(start_point[0], end_point[0]):
        for y in range(start_point[1], end_point[1]):
            total += mask[x, y] // 255
            n += 1
    average = total / n
    
    # cv2.imshow(label + " img", img)
    cv2.imshow(label + " res", res)
    
    return average
```

```python
capl = cv2.VideoCapture(2)
while 1:
    good, flip_img = capl.read()
    img = cv2.rotate(flip_img, cv2.ROTATE_180)
    if (good):
        print(
            orange := findColor(img, "orange", lower_orange, upper_orange),
#             findColor(img, "green", lower_green, upper_green, conversion=cv2.COLOR_BGR2YCgCo),
            pink := findColor(img, "pink", lower_pink, upper_pink),
        )
        if orange > 0.5: x = ("orange")
        elif pink > 0.5: x = ("pink")
        else: x = ("green")
        cv2.putText(img, x, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 2, 255)
        cv2.rectangle(img, start_point, end_point, (255, 0, 0), 2)
        cv2.imshow("img", img)
    if (cv2.waitKey(1) & 0xFF == ord('q')):
        break
capl.release()
cv2.destroyAllWindows()
```
